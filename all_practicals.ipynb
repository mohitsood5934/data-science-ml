```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score


df = pd.read_csv('./Downloads/ML/DataSets/Telco-Customer-Churn.csv')
print(df.head(5))
```

```python
print(df.info())
#EDA 
```

```python
# converting TotalCharges column to numeric as it was object
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors = 'coerce')

print(df.info())
```

```python
df.dropna(how='any',inplace = True)
print(len(df))
```

```python
# calculating churned rate
df.Churn.value_counts() / len(df) * 100
```

```python
# splitting in to X, y dataset 
X = df.drop(['customerID','Churn'], axis=1)
y = df.Churn.values
```

```python
X= pd.get_dummies(X,columns= ['gender','Partner', 'Dependents', 'PhoneService','MultipleLines','InternetService','OnlineSecurity',
                              'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract',
                              'PaperlessBilling','PaymentMethod'
                             ], drop_first = True)
```

```python
# splitting the data in to training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
print(len(X_train))
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

print('Scaled data (test and train data sets)')
print(X_train_sc, X_test_sc)
```

```python
# call the KNN Classifier
from sklearn.neighbors import KNeighborsClassifier

# initiating the classifier
model = KNeighborsClassifier()

# passing the data to classifiers
model.fit(X_train_sc,y_train)

y_pred = model.predict(X_test_sc)
print(y_pred)
```

```python
# classfication metrics to check how model is performing
from sklearn.metrics import accuracy_score
print("KNN Model accuracy score", accuracy_score(y_test,y_pred)*100)
```

```python
# decision tree classifier
from sklearn.tree import DecisionTreeClassifier 

# initiating the classifier
model_dt = DecisionTreeClassifier()

# passing the data to the classifier
model_dt.fit(X_train_sc,y_train)
y_pred_dt = model_dt.predict(X_test_sc)
print(y_pred_dt)

print("Decision Tree Model accuracy score", accuracy_score(y_test,y_pred_dt)*100)
```

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier

# initiating the classifier
model_rf = RandomForestClassifier(n_estimators = 200)

# passing the data to the classifier
model_rf.fit(X_train_sc,y_train)
y_pred_rf = model_dt.predict(X_test_sc)
print(y_pred_rf)
```

```python
# classfication metrics to check how model is performing
from sklearn.metrics import accuracy_score
print("Random Forest Model accuracy score", accuracy_score(y_test,y_pred_rf)*100)
```

```python
from sklearn.metrics import recall_score, precision_score

# Fix: Add pos_label='Yes' (positive class = churn)
knn_recall = recall_score(y_test, y_pred, pos_label='Yes')
decisiontree_recall = recall_score(y_test, y_pred_dt, pos_label='Yes')
randomforest_recall = recall_score(y_test, y_pred_rf, pos_label='Yes')

print("KNN Recall:", knn_recall)
print("Decision Tree Recall:", decisiontree_recall)
print("Random Forest Recall:", randomforest_recall)
```

```python
from sklearn.metrics import classification_report

# Fix: Add pos_label='Yes' (positive class = churn)
knn_cr = classification_report(y_test, y_pred)
decisiontree_cr = classification_report(y_test, y_pred_dt)
randomforest_cr = classification_report(y_test, y_pred_rf)

print("KNN recall classification report", knn_cr)
print("Decision tree classification report", decisiontree_cr)
print("Random Forest classification report", randomforest_cr)
```
